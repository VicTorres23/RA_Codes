```python
import pandas as pd
import re

master_sequence = pd.read_excel("C:/Users/vemma/Documents/Master in Bioinformatics/Fall_2024_Semester/Graduate Research/MasterSequence_Database (06.15.24) (1).xlsx", sheet_name="To be added")

HCO = "TAAACTTCAGGGTGACCAAAAAATCA"
LCO = "GGTCAACAAATCATAAAGATATTGG"
ITS4 = "TCCTCCGCTTATTGATATGC"
ITS5 = "GGAAGTAAAAGTCGTAACAAGG"

for i in range(len(master_sequence["Genus"])):
    if master_sequence["Genus"][i] == "Hexarthra" or master_sequence["Genus"][i] == "Rhinoglena" or (master_sequence["Genus"][i] == "Plationus " and master_sequence["Species"][i] == "patulus"):
        if master_sequence["Gene"][i] == "COI" and (re.search(HCO, master_sequence["Sequence"][i]) or re.search(LCO, master_sequence["Sequence"][i])):
            print("_________________________________")
            print("Original sequence:\n" + str(master_sequence["Sequence"][i]) + "\n")
            print("HCO primer in the sequence: " + str(HCO in master_sequence["Sequence"][i]) + "\n")
            print("LCO primer in the sequence: " + str(LCO in master_sequence["Sequence"][i]) + "\n")
            master_sequence["Sequence"][i] = re.sub(r"[ATGC]?" + HCO, "", master_sequence["Sequence"][i])
            master_sequence["Sequence"][i] = re.sub(LCO, "", master_sequence["Sequence"][i])
            print("Cleaned sequence:\n" + str(master_sequence["Sequence"][i]))
            print("_________________________________")
        elif master_sequence["Gene"][i] == "ITS" and (re.search(ITS4, master_sequence["Sequence"][i]) or re.search(ITS5, master_sequence["Sequence"][i])):
            print("_________________________________")
            print("Original sequence:\n" + str(master_sequence["Sequence"][i]) + "\n")
            print("ITS4 primer in the sequence: " + str(ITS4 in master_sequence["Sequence"][i]) + "\n")
            print("ITS5 primer in the sequence: " + str(ITS5 in master_sequence["Sequence"][i]) + "\n")
            master_sequence["Sequence"][i] = re.sub(ITS4, "", master_sequence["Sequence"][i])
            master_sequence["Sequence"][i] = re.sub(ITS5, "", master_sequence["Sequence"][i])
            print("Cleaned sequence:\n" + str(master_sequence["Sequence"][i]))
            print("_________________________________")
        Info = ">" + master_sequence["Genus"][i] + "_" + str(master_sequence["Species"][i]) + "_" + str(master_sequence["Location"][i]) + "_" + str(master_sequence["Code5"][i]) + "_" + str(master_sequence["Gene"][i]) + "\n" + str(master_sequence["Sequence"][i])
        new_fasta = open(master_sequence["Genus"][i] + "_" + str(master_sequence["Species"][i]) + "_" + str(master_sequence["Location"][i]) + "_" + str(master_sequence["Code5"][i]) + "_" + str(master_sequence["Gene"][i]) + ".fa", "w")
        new_fasta.write(Info)
```
